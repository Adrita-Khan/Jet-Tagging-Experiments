```python
import os
import shutil

# Base path where all the part folders are
base_dir = '/media/drive1/jettag/Jet-Tagging/raqib-pcn-experiments/training-experiments/training_1M/data'

# List of part folders 
part_folders = [f'JetClass_Pythia_train_100M_part{i}' for i in range(10)]

# Get all class names (if you want to restrict or validate)
classes = ["HToBB", "HToCC", "HToGG", "HTolW2QL", "HTolW4Q", 
           "TTBar", "TTBarLep", "WToQQ", "ZJetsToNuNu", "ZToQQ"]

# Create folders and move files
for folder in part_folders:
    part_path = os.path.join(base_dir, folder)
    for filename in os.listdir(part_path):
        if not filename.endswith('.root'):
            continue
        class_name = filename.split('_')[0]
        source_file = os.path.join(part_path, filename)
        target_dir = os.path.join(base_dir, class_name)
        os.makedirs(target_dir, exist_ok=True)
        target_file = os.path.join(target_dir, filename)
        shutil.move(source_file, target_file)
        print(f"Moved {filename} to {class_name}/")

print("Done organizing files by class!")
```
